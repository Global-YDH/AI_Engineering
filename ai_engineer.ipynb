```python
import time
import random

start = time.time()
sum = 0
for i in range(1,100000001):
  random_number = random.random()
  sum += random_number

end = time.time()

print("[0,1] 범위의 실수 난수를 생성하여 1억번 반복했을 때의 걸리는 시간:", (end-start))
```
